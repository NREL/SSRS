## Focus on temp at 500 mb

```python
Hx = H.xarray("TMP:500 mb", remove_grib=False)
```

```python
Hx
```

```python
latc = Hx.coords['latitude']
lonc = Hx.coords['longitude']
```

```python
# Hx.t.loc[{'longitude': lonc[(lonc >= 237.280472) & (lonc <= 237.33381)]}]
latc_mask = (latc >= 41.6) & (latc <= 41.8)
lonc_mask = (lonc >= 287.23) & (lonc <= 287.24)
mask = latc_mask & lonc_mask
Hx.t.where(mask)
```

```python
Hx.t.where(mask).sum() / float(mask.sum())
```
